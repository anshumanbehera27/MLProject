```python
import numpy as np 
traning_data = np.loadtxt('0000000000002329_training_diabetes_x_y_train.csv',delimiter = ',')
test_data = np.loadtxt('0000000000002329_test_diabetes_x_test.csv',delimiter = ',')
print(traning_data.shape)
print(test_data.shape)
```

```python
# we have alread thr perdict data so 
x_test = test_data
x_train = traning_data[:,:10]
y_train = traning_data[:,10:]
print(x_test.shape)
print(x_train.shape)
print(y_train.shape) # test 
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg1.fit(x_train,y_train)
```

```python
m = alg1.coef_
c = alg1.intercept_
print('m',m,'c',c)
```

```python
y_pred = alg1.predict(x_test)
y_pred_decimal5 = np.around(y_pred,decimals = 5)
y_pred_decimal5
```

```python
np.savetxt('pred.csv',y_pred_decimal5,delimiter = ',',fmt = '%f')
```
